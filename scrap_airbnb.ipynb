```python
from urllib import request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
lieu='PARIS LA DEFENSE ARENA'.replace(' ', '-')
url="https://www.airbnb.fr/s/"+lieu+"/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&flexible_trip_lengths%5B%5D=one_week&price_filter_input_type=0&price_filter_num_nights=1&date_picker_type=calendar&checkin=2023-01-01&checkout=2022-01-02&source=structured_search_input_header&search_type=search_query"
request_text = request.urlopen(url).read()
htmlpage= BeautifulSoup(request_text, "html")
```

```python
hotels=[]
for item in htmlpage.findAll('div',{'class':'g1qv1ctd cb4nyux dir dir-ltr'}):
    hotel={}
    hotel['Titre']=item.find('div',{'class':'t1jojoys dir dir-ltr'}).text.replace('\n','')
    hotel['Type']=item.find('span',{'class':'t6mzqp7 dir dir-ltr'}).text.replace('\n','')
    hotel['Lit']=item.find('div',{'class':'f15liw5s s1cjsi4j dir dir-ltr'}).text.replace('\n','')
    hotel['Prix (€)']=item.find('div',{'class':'_tt122m'}).text.replace('\n','').replace('€ au total','')
    item_rating=item.find('span',{'class':'r1dxllyb dir dir-ltr'})
    if(item_rating != None):
        rating= item_rating.text.replace(')','').split('(')
        if(len(rating)==2):
            hotel['Note (sur 5)']=rating[0]
            hotel['Nombre de revues']=rating[1]
    hotels.append(hotel)
```

```python
df= pd.DataFrame(hotels)
df
```
